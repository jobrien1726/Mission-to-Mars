```python
#Dependencies
from splinter import Browser
from bs4 import BeautifulSoup as bs
import requests
```

# NASA Mars News

```python
# Chrome Driver, https://splinter.readthedocs.io/en/latest/drivers/chrome.html
!which chromedriver
```

```python
executable_path = {'executable_path': '/usr/local/bin/chromedriver'}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
# URL of NASA Mars News Site to be scraped
news_url = "https://mars.nasa.gov/news/?page=0&per_page=40&order=publish_date+desc%2Ccreated_at+desc&search=&category=19%2C165%2C184%2C204&blank_scope=Latest"
browser.visit(news_url)
```

```python
# HTML object
html = browser.html

# Parse HTML with Beautiful Soup
soup = bs(html, "html.parser")
```

```python
# Collect the latest News Title and Paragraph Text
news_title = soup.find("div", class_= "content_title").a.text.strip()
print(news_title)

news_p = soup.find("div", class_= "article_teaser_body").text.strip()
print(news_p)
```

```python
browser.quit()
```

# Mars Weather

```python
import requests
```

```python
# Twitter URL for Mars weather tweet
twitter_url = "https://twitter.com/marswxreport?lang=en"
```

```python
# I was initially using splinter, but browser was not allowing me to access the text 
# from the tweet (kept returning "Log in") so I switched to requests
response = requests.get(twitter_url)

# Create BeautifulSoup object; parse with 'html.parser'
soup = bs(response.text, 'html.parser')

print(soup.prettify())
```

```python
# Scrape the latest Mars weather tweet
mars_weather = soup.find("p", class_="TweetTextSize TweetTextSize--normal js-tweet-text tweet-text").text.strip()
print(mars_weather)
```

# Mars Facts

```python
import pandas as pd
```

```python
# Space Facts URL for Mars Facts
facts_url = "https://space-facts.com/mars/"
```

```python
# Use read_html to scrape table from Space Facts site
mars_facts = pd.read_html(facts_url)
mars_facts
```

```python
# Save table as a pandas dataframe, rename columns and set index
mars_facts_df = mars_facts[0]
mars_facts_df = mars_facts_df.rename(columns={0:"planet_metrics", 1:"mars"})
mars_facts_df.set_index("planet_metrics", inplace=True)
mars_facts_df
```

```python
# Convert dataframe to html table string and save as html file
mars_html_table = mars_facts_df.to_html()
mars_html_table
```

```python
# Get rid of \n
mars_html_table.replace('\n', '')
```

```python
# Save mars table to html file 
mars_facts_df.to_html("mars_table.html")
```

# Mars Hemispheres

```python
# Chrome Driver, https://splinter.readthedocs.io/en/latest/drivers/chrome.html
!which chromedriver
```

```python
executable_path = {'executable_path': '/usr/local/bin/chromedriver'}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
# Hemisphere List Page
hemi_url = "https://astrogeology.usgs.gov/search/results?q=hemisphere+enhanced&k1=target&v1=Mars"
browser.visit(hemi_url)
```

```python
# HTML object
html = browser.html

# Parse HTML with Beautiful Soup
soup = bs(html, "html.parser")
```

```python
#Find all hemisphere titles
results = soup.find_all("div", class_="description")
results
```

```python
#Create a list of all the Hemisphere Titles from the first page
hemi_titles = []

for result in results:
    hemi_title = result.h3.text
    hemi_titles.append(hemi_title)
    
hemi_titles
```

```python
#Base url to be used later for full image url
base_img_url = "https://astrogeology.usgs.gov"
```

```python
# Find image urls for each hemisphere
hemi_img_urls = []

for title in hemi_titles:
    # Format to put into url
    title = title.lower().replace(" hemisphere ", "_")
   
    # Navigate to page for each Hemisphere
    indiv_pg_url = f"https://astrogeology.usgs.gov/search/map/Mars/Viking/{title}"
    browser.visit(indiv_pg_url)

    # HTML object
    html = browser.html

    # Parse HTML with Beautiful Soup
    soup = bs(html, "html.parser")

    # Find full size image url
    img_url = soup.find("img", class_="wide-image")["src"]
    #print(img_url)

    full_img_url = base_url + img_url
    #print(full_img_url)
    
    # Append image url to list of hemisphere image urls
    hemi_img_urls.append(full_img_url)
    
hemi_img_urls
```

```python
# Save hemisphere titles and image urls to a dictionary
hemisphere_image_urls = []

for x in range(4):
    hemi_img_dict = {"title": hemi_titles[x], "img_url": hemi_img_urls[x]}
    hemisphere_image_urls.append(hemi_img_dict)
    
hemisphere_image_urls
```

```python
browser.quit()
```
